```python
monthly_view = dfb2.groupby(dfb2['Date'].dt.month).count()
monthly_view
```

```python
monthly_per_year_view = dfb2.groupby([dfb2['Date'].dt.month,dfb2['Date'].dt.year]).count()
monthly_per_year_view
```

```python
monthly_per_year_view.index.names=['Month','Year']
monthly_per_year_view.head()
```

```python
monthly_per_year_view['Date'].plot.line(figsize=(14,5))
p.show()
```

```python
unstacked_month_year=monthly_per_year_view['Date'].unstack()
unstacked_month_year
```

```python
p.style.use('seaborn-dark-palette')
unstacked_month_year.plot.line(figsize=(16,5))
p.legend(loc=3)
```

```python
transposed_ym = unstacked_month_year.transpose()
transposed_ym
```

```python
stdev=transposed_ym.std()
mean=transposed_ym.mean()
stdev,mean
```

```python
mean.plot.line(figsize=(16,5),yerr=stdev)
```

```python
year_range=pd.date_range('2013-01-01', periods=12,freq='%m')
year_range
```

```python
daysinmonth=year_range.to_series().dt.daysinmonth
daysinmonth
```

```python
daysinmonth.index=year_range.to_series().dt.month
daysinmonth
```

```python
cols=unstacked_month_year.columns
weightedaveragemy=pd.DataFrame()
for i in cols:
    weightedaveragemy[i]=unstacked_month_year[i]/daysinmonth
weightedaveragemy
```

```python
weightedaveragemy.plot.line(figsize=(16,5))
p.legend(loc=3)
unstacked_month_year.plot.line(figsize=(16,5))
p.legend(loc=3)
```

```python
weightedaverageym=weightedaveragemy.transpose()
weightedaverageym.mean().plot.line(figsize=(16,5),yerr=weightedaverageym.std())
p.show()
mean.plot.line(figsize=(16,5),yerr=stdev)
p.show()
```
